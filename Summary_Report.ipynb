```python
import PyPDF2
```

```python
pdf_path = '/content/1684756281533.pdf'
```

```python
pdf_path
```

```python
with open(pdf_path, 'rb') as file:
    # Create a PDF reader object
    reader = PyPDF2.PdfReader(file)
    
    # Initialize a variable to store the accumulated text
    accumulated_text = ""
    
    # Iterate over the pages
    for page_number in range(len(reader.pages)):
        # Get the specific page
        page = reader.pages[page_number]
        
        # Extract the text from the page
        text = page.extract_text()
        
        # Add the text to the accumulated text
        accumulated_text += text
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer
```

```python
# Specify the language for summarization
language = "english"

# Initialize the LexRank summarizer
summarizer = LexRankSummarizer()
```

```python
import nltk
nltk.download('punkt')
```

```python
parser = PlaintextParser.from_string(accumulated_text, Tokenizer(language))
```

```python
num_sentences = 15
```

```python
summary = summarizer(parser.document, num_sentences)
```

```python
    print("Summarized Content:")
    for sentence in summary:
        print(sentence)
```
